```python
SPLIT = floor(len(data)*0.6)
spend_data = data.copy()

new_df = spend_data[date + media_spend + week  + original_target + target + log_target].reset_index(drop=True)


train_df = new_df[date + original_target + target + log_target + media_spend + week][:SPLIT].reset_index(drop=True)
test_data = new_df[date + original_target + target + log_target + media_spend + week ][SPLIT:].reset_index(drop=True)
```

```python
from functools import partial
import optuna as opt
from xgboost import XGBRegressor, Booster
from sklearn.model_selection import train_test_split
```

```python
def optuna_objective(trial, data, target):

    # xgb params
    n_estimators = trial.suggest_int('n_estimators', 100, 1000)
    max_depth = trial.suggest_int('max_depth', 2, 10)
    subsample  = trial.suggest_float('subsample', 0.6, 1)
    learning_rate = trial.suggest_float('learning_rate', 0.01, 0.1)
    objective = trial.suggest_categorical('objecctive',['reg:squarederror', 'reg:squaredlogerror'])
    child_weight = trial.suggest_int('min_child_weight', 1, 5)

    params = {'n_estimators':n_estimators,
              'max_depth':max_depth,
              'subsample':subsample,
              'learning_rate':learning_rate,
              'min_child_weight':child_weight,
            #   'lambda':trial.suggest_float('lambda', 0, 1),
            #   'alpha':trial.suggest_float('alpha', 0, 1),
              'objective':objective}

    # train
    scores= []

    xgb = XGBRegressor(seed=42, **params)

    x_train, x_test, y_train, y_test = train_test_split(data[media_spend + week ].values,
                                      data[target].values,
                                      test_size=0.33, random_state=42)
    x_test, x_val, y_test, y_val = train_test_split(x_test, y_test, test_size=0.3, random_state=42)
    

    xgb.fit(x_train, y_train, eval_metric='rmse', eval_set=[(x_val, y_val)], early_stopping_rounds=10, verbose=False)
    pred = xgb.predict(x_test)

    rmse = np.sqrt(mean_squared_error(y_true=y_test, y_pred=pred))
    scores.append(rmse)

    trial.set_user_attr('scores', scores)
    trial.set_user_attr('params', params)
    
    return np.mean(scores)



def optimize(n_trials, data:pd.DataFrame, target):
    opt.logging.set_verbosity(opt.logging.WARNING)
    study = opt.create_study(direction='minimize', sampler=opt.samplers.TPESampler())
    objective_func = partial(optuna_objective, data=data, target=target)
    study.optimize(objective_func, n_trials=n_trials, show_progress_bar=True)
    
    return study
```

```python
TRIALS = 500

experiment1 = optimize(TRIALS, train_df, original_target)      # target = y
experiment2 = optimize(TRIALS, train_df, target)               # target = y_logdiff
experiment3 = optimize(TRIALS, train_df, log_target)           # target = y_log


xgb_params1 = experiment1.best_trial.user_attrs['params']
xgb_params2 = experiment2.best_trial.user_attrs['params']
xgb_params3 = experiment3.best_trial.user_attrs['params']
```

```python
def spend_effect(data:pd.DataFrame, shap_values:np.ndarray):
    shap_values_df = pd.DataFrame(shap_values, columns=[media_spend + week ])
    total_effect = shap_values_df.abs().sum(axis=0)
    effect_share_df = pd.DataFrame(total_effect / total_effect.sum()).reset_index(drop=False)
    effect_share_df.columns = ['feature', 'share']


    total_spend = data[media_spend].sum(axis=0)

    spend_share_df = pd.DataFrame(total_spend / total_spend.sum()).reset_index(drop=False)
    spend_share_df.columns = ['feature', 'share']


    spend_effect_share = pd.DataFrame({})
    spend_effect_share['feature'] = effect_share_df['feature']
    spend_effect_share['spend share'] = spend_share_df['share']
    spend_effect_share['effect share'] = effect_share_df['share']

    if len(effect_share_df) > len(spend_share_df):
        diff = len(effect_share_df) - len(spend_share_df)
        for i in range(diff):
            spend_effect_share.loc[len(spend_effect_share)-(i+1), 'spend share'] = 0

    return spend_effect_share
```

```python
spend_effect_df = spend_effect(data, shap_values)
spend_effect_df.plot(x='feature', y=['spend share', 'effect share'], kind='bar', figsize=(10,5))
```

```python
def plot_importance(data, shap_values):
    corr_list = []
    for i, feature in enumerate(media_spend + week ):
        corr = np.corrcoef(shap_values[:, i], data[feature].values)[0,1]
        corr_list.append(corr)

    corr_df = pd.DataFrame({})
    corr_df['feature'] = media_spend + week 
    corr_df['corr'] = corr_list

    sign = np.where(corr_df['corr']>0,'red','blue')
    shap_abs = np.mean(np.abs(shap_values), axis=0)
    corr_df['shap'] = shap_abs
    corr_df.sort_values(by='shap', ascending=True, inplace=True)

    corr_df.plot(x='feature', y='shap', kind='bar')

plot_importance(data, shap_values)
```

```python
def shap_feature_importance(shap_values, data, figsize = (10, 5)):
    
    feature_list = data.columns
    
    if not isinstance(shap_values, pd.DataFrame):
        shap_v = pd.DataFrame(shap_values)
        shap_v.columns = feature_list
    else:
        shap_v = shap_values
    
        
    df_v = data.copy().reset_index().drop('index',axis=1)
    
    # Determine the correlation in order to plot with different colors
    corr_list = list()
    for i in feature_list:
        corr = np.corrcoef(shap_v[i],df_v[i])[0,1]
        corr_list.append(corr)
    corr_df = pd.concat([pd.Series(feature_list),pd.Series(corr_list)],axis=1).fillna(0)

    corr_df.columns  = ['Variable','Corr']
    corr_df['Sign'] = np.where(corr_df['Corr']>0,'blue','red')
    
    # Plot it
    shap_abs = np.abs(shap_v)
    k=pd.DataFrame(shap_abs.mean()).reset_index()
    k.columns = ['Variable','SHAP_abs']
    k2 = k.merge(corr_df,left_on = 'Variable',right_on='Variable',how='inner')
    k2 = k2.sort_values(by='SHAP_abs',ascending = True)
    colorlist = k2['Sign']
    ax = k2.plot.barh(x='Variable',y='SHAP_abs',color = colorlist, figsize=figsize,legend=False)
    ax.set_xlabel("SHAP Value (Blue = Positive Impact)")


shap_feature_importance(shap_values, data[media_spend + week ])
```
